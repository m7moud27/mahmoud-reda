```
training_data = []

def train_data():
    for clase in classes:

        path = os.path.join(data,clase)
        class_num =classes.index(clase)
        
        for imag in tqdm(os.listdir(path)):
            try:
                imag_array = cv2.imread(os.path.join(path,imag),cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(imag_array,(Imag_size,Imag_size))
                training_data .append([new_array,class_num])

            except Exception as e:
                pass
            
            except OSError as e:
                print('OSErrroBad img most likely', e, os.path.join(path, imag))

            except Exception as e:
                print('general exception', e, os.path.join(path,imag))
train_data()
```
